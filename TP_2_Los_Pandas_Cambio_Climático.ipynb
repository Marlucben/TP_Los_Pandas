#2. Calcular las temperaturas máximas y mínimas promedio para cada ciudad donde hay registro detallado y cada mes del último año.


```python
def tempMaxYMinPromedioPorCiudadPorMes(df:pd.DataFrame)->pd.DataFrame:
 # Crear una copia del DataFrame y extraer el mes sin modificar el original
 df_mes = df.copy()
 df_mes["mes"] = df_mes["fecha"].dt.month

 # Agrupar por nombre de ciudad y mes, calcular promedios
 ultimo_año_promedio = df_mes.groupby(["nombre", "mes"])[["temp_min", "temp_max"]].mean().reset_index()

 # Calcular el promedio entre min y max
 ultimo_año_promedio["temp_prome"] = (ultimo_año_promedio["temp_min"] + ultimo_año_promedio["temp_max"]) / 2
 #Convertimos los datos a entero
 ultimo_año_promedio["mes"] = ultimo_año_promedio["mes"].astype(int)

 # Redondeamos las temperaturas a un decimal
 ultimo_año_promedio["temp_min"] = ultimo_año_promedio["temp_min"].round(1)
 ultimo_año_promedio["temp_max"] = ultimo_año_promedio["temp_max"].round(1)
 ultimo_año_promedio["temp_prome"] = ultimo_año_promedio["temp_prome"].round(1)

 return ultimo_año_promedio
```

Funcion convertidor de columnas largas a corta para el df historico:

```python
def normalizarLetras(df:pd.DataFrame,columna)->pd.DataFrame:
      df[columna] = (
        df[columna]
          .str.strip()
          .str.replace('Á', 'A', regex=False)
          .str.replace('É', 'E', regex=False)
          .str.replace('Í', 'I', regex=False)
          .str.replace('Ó', 'O', regex=False)
          .str.replace('Ú', 'U', regex=False)
          .str.replace('á', 'a', regex=False)
          .str.replace('ó', 'o', regex=False)
          .str.replace('í', 'i', regex=False)
          .str.replace('é', 'e', regex=False)
          .str.replace('ú', 'u', regex=False)
          .str.replace('°', '', regex=False)
          .str.replace(r'\*$', '', regex=True)
          .str.replace(r'\sAERO$', '', regex=True)

          .str.upper())

def convertir_historico(df_hist:pd.DataFrame)->pd.DataFrame:
 df = df_hist.copy()#Copiamos el dataframa para no modificarlo


 meses_num = list(range(1, 13))

 #Recortamos solo columnas útiles
 df_corto = df.iloc[:, [0, 1] + list(range(2, 14))].copy()
 df_corto.columns = ['nombre', 'variable'] + meses_num

 #Pasar a formato largo
 df_largo = df_corto.melt(id_vars=['nombre', 'variable'],
                             var_name='mes',
                             value_name='valor')

 # corrije y convierte todos los valores a numeros para evitar errores de tipados
 df_largo['valor'] = pd.to_numeric(df_largo['valor'], errors='coerce')
 #Pivotear para tener columnas por tipo de temperatura
 df_pivot = df_largo.pivot_table(index=['nombre', 'mes'],
                                    columns='variable',
                                    values='valor').reset_index()

 #Renombrar columnas (opcional, si los valores son exactos)
 df_pivot.columns.name = None
 df_pivot = df_pivot.rename(columns={
        'Temperatura mínima (°C)': 'temp_min',
        'Temperatura máxima (°C)': 'temp_max',
        'Temperatura (°C)': 'temp_prome'})

 #Ordenar
 df_pivot = df_pivot.sort_values(['nombre', 'mes'])
 normalizarLetras(df_pivot,'nombre')#normalizamos todas las letras y simbolos


 return df_pivot[['nombre', 'mes', 'temp_min', 'temp_max', 'temp_prome']]
```

```python
def crearTablaDeDiferencias(tabla):
  tabla["dif_maximas"] = tabla["temp_max_actual"] - tabla["temp_max_hist"]
  tabla["dif_minimas"] = tabla["temp_min_actual"] - tabla["temp_min_hist"]
  tabla["dif_promedio"] = tabla["temp_prome_actual"] - tabla["temp_prome_hist"]
  tabla_prom = tabla.groupby("nombre")[["dif_maximas", "dif_minimas", "dif_promedio"]].mean().reset_index()
  tabla_prom["dif_maximas"] = tabla_prom["dif_maximas"].round(2)
  tabla_prom["dif_minimas"] = tabla_prom["dif_minimas"].round(2)
  tabla_prom["dif_promedio"] = tabla_prom["dif_promedio"].round(2)
  return tabla_prom

def sortTabla(tabla, value):
  tablaSorted = tabla.sort_values(value, ascending=False)
  return tablaSorted

tabla_diferencias=crearTablaDeDiferencias(tabla_combinada)
ranking_maximas = sortTabla(tabla_diferencias, "dif_maximas")
ranking_minimas = sortTabla(tabla_diferencias, "dif_minimas")
ranking_promedios = sortTabla(tabla_diferencias, "dif_promedio")
ranking_maximas, ranking_minimas, ranking_promedios
```

```python
def extremos_termicos(df_max, df_min):
    extremos = pd.merge(df_max, df_min, on='Cuidad')

    extremos['Extremos_Termicos'] = extremos['Maxima'] - extremos['Minima']

    extremos.columns = ['Cuidad', 'Maxima_Verano', 'Minima_Invierno', 'Extremos_Termicos']

    return extremos
```

```python
def ciudades_con_extremos(df_max_hist, df_min_hist, df_max_actual, df_min_actual, top_n=5):
    # Renombrar columnas para evitar conflictos
    df_max_hist_renamed = df_max_hist.rename(columns={"Maxima": "Maxima_Verano_Hist"})
    df_min_hist_renamed = df_min_hist.rename(columns={"Minima": "Minima_Invierno_Hist"})
    df_max_actual_renamed = df_max_actual.rename(columns={"Maxima": "Maxima_Verano_Actual"})
    df_min_actual_renamed = df_min_actual.rename(columns={"Minima": "Minima_Invierno_Actual"})

    # Normalizar nombres de ciudades
    for df in [df_max_hist_renamed, df_min_hist_renamed, df_max_actual_renamed, df_min_actual_renamed]:
        df["Cuidad"] = df["Cuidad"].str.strip().str.upper()

    # Unir por "Cuidad"
    extremos = df_max_hist_renamed.merge(df_min_hist_renamed, on="Cuidad", how="inner")
    extremos = extremos.merge(df_max_actual_renamed, on="Cuidad", how="inner")
    extremos = extremos.merge(df_min_actual_renamed, on="Cuidad", how="inner")

    # Calcular extremos térmicos
    extremos["Extremos_Termicos_Hist"] = extremos["Maxima_Verano_Hist"] - extremos["Minima_Invierno_Hist"]
    extremos["Extremos_Termicos_Actual"] = extremos["Maxima_Verano_Actual"] - extremos["Minima_Invierno_Actual"]

    # Ordenar y quedarnos con top N
    extremos_ordenados = extremos.sort_values(by="Extremos_Termicos_Hist", ascending=False).head(top_n)

    # Redondear valores
    for col in ["Maxima_Verano_Hist", "Minima_Invierno_Hist", "Maxima_Verano_Actual", "Minima_Invierno_Actual", "Extremos_Termicos_Hist", "Extremos_Termicos_Actual"]:
        extremos_ordenados[col] = extremos_ordenados[col].round(1)

    return extremos_ordenados
```

#4.b Ciudades donde se acentuaron los extremos (temperaturas máximas mas elevadas en verano y temperaturas mínimas más bajas en invierno)

```python
def extremos_ciudad(df_max_hist, df_min_hist, df_max_actual, df_min_actual, ciudad_buscada, graficar=True):
    # Renombrar columnas
    df_max_hist_renamed = df_max_hist.rename(columns={"Maxima": "Maxima_Verano_Hist"})
    df_min_hist_renamed = df_min_hist.rename(columns={"Minima": "Minima_Invierno_Hist"})
    df_max_actual_renamed = df_max_actual.rename(columns={"Maxima": "Maxima_Verano_Actual"})
    df_min_actual_renamed = df_min_actual.rename(columns={"Minima": "Minima_Invierno_Actual"})

    # Normalizar nombres de ciudades
    for df in [df_max_hist_renamed, df_min_hist_renamed, df_max_actual_renamed, df_min_actual_renamed]:
        df["Cuidad"] = df["Cuidad"].str.strip().str.upper()

    ciudad_buscada = ciudad_buscada.strip().upper()

    # Filtrar la ciudad buscada
    try:
        datos = df_max_hist_renamed[df_max_hist_renamed["Cuidad"] == ciudad_buscada].merge(
            df_min_hist_renamed[df_min_hist_renamed["Cuidad"] == ciudad_buscada], on="Cuidad"
        ).merge(
            df_max_actual_renamed[df_max_actual_renamed["Cuidad"] == ciudad_buscada], on="Cuidad"
        ).merge(
            df_min_actual_renamed[df_min_actual_renamed["Cuidad"] == ciudad_buscada], on="Cuidad"
        )
    except:
        print("No se encontraron datos suficientes para esa ciudad.")
        return None

    # Calcular extremos térmicos
    datos["Extremos_Termicos_Hist"] = datos["Maxima_Verano_Hist"] - datos["Minima_Invierno_Hist"]
    datos["Extremos_Termicos_Actual"] = datos["Maxima_Verano_Actual"] - datos["Minima_Invierno_Actual"]

    # Redondear
    datos = datos.round(1)

    if graficar:
        # Valores
        ciudad = datos.iloc[0]["Cuidad"]
        max_hist = datos.iloc[0]["Maxima_Verano_Hist"]
        min_hist = datos.iloc[0]["Minima_Invierno_Hist"]
        max_actual = datos.iloc[0]["Maxima_Verano_Actual"]
        min_actual = datos.iloc[0]["Minima_Invierno_Actual"]
        extremo_hist = datos.iloc[0]["Extremos_Termicos_Hist"]
        extremo_actual = datos.iloc[0]["Extremos_Termicos_Actual"]

        x = np.array([0, 1])  # 0 para histórico, 1 para actual

        fig, ax = plt.subplots(figsize=(8, 6))

        # Barras principales: extremos térmicos
        ancho = 0.1
        ax.bar(x, [extremo_hist, extremo_actual], width=ancho, color=["#82e0aa", "#82e0aa"], label="Extremos Térmicos")

        # Barras pequeñas: temperaturas máximas y mínimas
        ax.bar(x - 0.15, [max_hist, max_actual], width=0.1, color="#e74c3c", label="T. Máxima")  # Rojo pastel
        ax.bar(x + 0.15, [min_hist, min_actual], width=0.1, color="skyblue", label="T. Mínima")

        # Etiquetas y detalles
        ax.set_xticks(x)
        ax.set_xticklabels(["Histórico", "Actual"])
        ax.set_ylabel("Temperatura (°C)")
        ax.set_title(f"Extremos Térmicos en {ciudad.title()}")
        ax.legend()

        # Agregar los valores encima de las barras
        valores = [extremo_hist, extremo_actual, max_hist, max_actual, min_hist, min_actual]
        posiciones = list(x) + list(x - 0.15) + list(x + 0.15)
        for pos, val in zip(posiciones, valores[0:6]):
            ax.annotate(f"{val:.1f}", (pos, val + 0.3), ha='center', fontsize=9)

        plt.grid(axis='y', linestyle='--', alpha=0.4)
        plt.tight_layout()
        plt.show()

    return datos
```
